## Step 1: Import the CSV file and clean it, then we input those data into a dictionary

```python
import csv

#create a dictionary class
class my_dict(dict):
  
    # __init__ function
    def __init__(self):
        self = dict()
          
    # Function to add key:value
    def add(self, key, value):
        self[key] = value
    
    # Function to get values from certain key ~ passed in
    def geT(self, key):
        return self.get(key)   


#creating the dictinaries for our 5 clothing stores
H_M = my_dict()
Zara = my_dict()
Forever_21 = my_dict()
Uniqlo = my_dict()
Shein = my_dict()

with open('Clothes_size.csv') as csv_file:
    #open the csv folder
    csv_reader = csv.reader(csv_file, delimiter=',')
    # line_num will be keeping track of the ur
    line_num = 0
    
    # parsing through the csv file and input in the data to our dictionary
    """
    Dictionary format:
    Key: the type of size (chest size, waist size or hip size)
    Value: the actual measurements
    """
    for row in csv_reader:
        if line_num < 6:    
            if line_num > 2:
                H_M.add(row[0], row[1:])      
        elif line_num < 12:
            if line_num > 8:
                Zara.add(row[0], row[1:]) 
        elif line_num < 18:
            if line_num > 14:
                Uniqlo.add(row[0], row[1:]) 
        elif line_num < 24:
            if line_num > 20:
                Shein.add(row[0], row[1:]) 
        elif line_num < 30:
             if line_num > 26:
                Forever_21.add(row[0], row[1:]) 
        line_num += 1
        
    #print(H_M)
    #print(Zara)
    #print(Uniqlo)
    #print(Shein)
    #print(Forever_21)
```

### Pre-MVP
This is our most simple MVP. We predict the user's size by using case studies about ectomorph, endomorph and mesomorph. There exist a simple formula for these 3 body types according to a collection of data points we found online

```python
import pandas as pd
# for a simple GUI
import tkinter as tk

root= tk.Tk()
canvas1 = tk.Canvas(root, width = 400, height = 300)
canvas1.pack()

Title = tk.Label(root, text= "Size Predictor", font=('Century 20 bold'))
canvas1.create_window(200, 30, window=Title)

entry1 = tk.Entry (root) 
canvas1.create_window(200, 80, window=entry1)

entry2 = tk.Entry (root) 
canvas1.create_window(200, 160, window=entry2)


def prediction():
    res_height = 215
    
    try:
        h = float(entry1.get())
        w = float(entry2.get())
    except:
        error = tk.Label(root, text= "Please enter in both height and weight!", font=('Century 13 bold'))
        canvas1.create_window(200, res_height, window=error) 
        return


    if w < 38 or w > 170 or  h < 110 or h > 220:
        error = tk.Label(root, text= "Error in input data. Check your input!", font=('Century 13 bold'))
        canvas1.create_window(200, res_height, window=error)
        pass
    else:
        R = w/((h/100)**3)
        if R < 12.8:
            body_type = 'ectomorph'
            chest_w = int(h*0.5)
            waist_w = int(h*0.35)
            hip_w = int(h*0.52)
        if R >= 12.8 and R < 14.6:
            body_type = 'mesomorph'
            chest_w = int(h*0.525)
            waist_w = int(h*0.375)
            hip_w = int(h*0.56)
        if R >= 14.6:
            body_type = 'endomorph'
            chest_w = int(h*0.55)
            waist_w = int(h*0.4)
            hip_w = int(h*0.6)

    counter = 0
    for shops in size_dict:
        for size in size_dict.get(shops):
            chest = size_dict.get(shops).get(size)[0]
            waist = size_dict.get(shops).get(size)[1]
            hip = size_dict.get(shops).get(size)[2]

            if chest_w > chest:
                pass
            else:
                if waist_w > waist:
                    pass
                else:
                    if hip_w > hip:
                        pass
                    else:
                        i = counter+1
                        res = str(i) + '.  ' + shops + ' size: ' + size
                        output = tk.Label(root, text= res, font=('Century 13 bold'))
                        canvas1.create_window(200, res_height, window=output)
                        res_height += 20
                        print(str(i) + '. ', shops, 'size:', size)
                        counter += 1
                        break  

button1 = tk.Button(text='Enter your height in cm')
canvas1.create_window(200, 110, window=button1)

button2 = tk.Button(text='Enter your weight in kg', command=prediction)
canvas1.create_window(200, 190, window=button2)

root.mainloop()
```

## Step 3: Recommend the user's size according to our dictionary

```python
import numpy as np

# getting the different sizes for H&M
HM_chest_sizes = H_M.geT('Chest_size')
HM_waist_sizes = H_M.geT('Waist_size')
HM_hip_sizes = H_M.geT('Hip_size')

# asking the user to input our size in
chest = int(input("What is your chest size? "))
waist = int(input("What is your waist size? "))
hip = int(input("What is your hip size? "))
""" 
result array that will be storing our indicies. For instance, inside of our CSV file, XS
has the index 2, S has the index 2 and 3 (since there are 2 type of S), etc...
""" 
res = []

def recommend(num, array):   
    index = 0
    for size in array: 
        try:
            """
            if our code goes here, the size is in the from a-b where a is the lower bound
            and b is the upper bound. By using .split, the 'curr' variable will be an array
            of our 2 bound number
            """
            curr = size.split("-")
            # convert from a string array to a number array
            int_array = [int(numeric_string) for numeric_string in curr]

            if num >= int_array[0] and num <= int_array[1]:
                # appending our current index to our result array
                res.append(index)
                #print(int_array)
                break
        except:
            """
            if our code goes here, our sizes are NOT in a range. Therefore, it is just a 
            sin
            """
            # if in this case, our size is not in a range
            pass
        index += 1

# to find the size of the user, I simple average out the 3 indices inside our array
def find_avg(array):
    total = 0
    for i in array:
        total += i
    return total/len(array)
        
recommend(chest, HM_chest_sizes)
recommend(waist, HM_waist_sizes)
recommend(hip, HM_hip_sizes)

avg = find_avg(res)

if avg >= 0 and avg <= 1:
    size = 'XS'
if avg >= 1 and avg <= 2:
    size = 'S'
elif avg >= 2 and avg <= 3:
    size = 'M'
elif avg >= 3 and avg <= 4:
    size = 'L'
else:
    size = 'XL'
print('For H&M, we recommend you size: {}'.format(size))
```
